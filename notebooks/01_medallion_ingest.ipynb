```python
import pandas as pd

tickets_pd = pd.read_csv("/Workspace/Users/nicholai.gay@gmail.com/tickets.csv")
faq_pd = pd.read_csv("/Workspace/Users/nicholai.gay@gmail.com/faq.csv")

display(tickets_pd)
display(faq_pd)
```

```python
tickets_df = spark.createDataFrame(tickets_pd)
faq_df = spark.createDataFrame(faq_pd)

display(tickets_df)
display(faq_df)
```

```python
spark.sql("SHOW CATALOGS").show(truncate=False)
```

```python
spark.sql("CREATE SCHEMA IF NOT EXISTS workspace.support_demo")
```

```python
(
    tickets_df.write
    .mode("overwrite")
    .format("delta")
    .saveAsTable("workspace.support_demo.bronze_tickets")
)

(
    faq_df.write
    .mode("overwrite")
    .format("delta")
    .saveAsTable("workspace.support_demo.bronze_faq")
)
```

```python
spark.sql("SHOW TABLES IN workspace.support_demo").show(truncate=False)
```

```python
display(spark.table("workspace.support_demo.bronze_tickets"))
display(spark.table("workspace.support_demo.bronze_faq"))
```

```python
from pyspark.sql import functions as F

silver_tickets_df = (
    spark.table("workspace.support_demo.bronze_tickets")
    .withColumn("created_at", F.to_date("created_at"))
    .withColumn("priority", F.upper(F.col("priority")))
    .withColumn("status", F.upper(F.col("status")))
    .withColumn("issue_type", F.initcap(F.col("issue_type")))
    .withColumn("product", F.trim(F.col("product")))
    .withColumn("customer_message", F.trim(F.col("customer_message")))
)

display(silver_tickets_df)
```

```python
silver_faq_df = (
    spark.table("workspace.support_demo.bronze_faq")
    .withColumn("title", F.trim(F.col("title")))
    .withColumn("content", F.trim(F.col("content")))
)

display(silver_faq_df)
```

```python
(
    silver_tickets_df.write
    .mode("overwrite")
    .format("delta")
    .saveAsTable("workspace.support_demo.silver_tickets")
)

(
    silver_faq_df.write
    .mode("overwrite")
    .format("delta")
    .saveAsTable("workspace.support_demo.silver_faq")
)
```

```python
spark.sql("SHOW TABLES IN workspace.support_demo").show(truncate=False)
```

```python
gold_issue_summary_df = (
    spark.table("workspace.support_demo.silver_tickets")
    .groupBy("product", "issue_type", "status")
    .count()
    .withColumnRenamed("count", "ticket_count")
    .orderBy("ticket_count", ascending=False)
)

display(gold_issue_summary_df)
```

```python
(
    gold_issue_summary_df.write
    .mode("overwrite")
    .format("delta")
    .saveAsTable("workspace.support_demo.gold_issue_summary")
)
```

```python
spark.sql("SHOW TABLES IN workspace.support_demo").show(truncate=False)
```

```python
display(spark.table("workspace.support_demo.gold_issue_summary"))
```
